Calculate the sea, road and total transport costs.

### Prepare workstation and read excel file

```python
# Import libraries
import pandas as pd
import numpy as np
```

```python
# Read the Excel file
transport_costs = pd.read_excel('transport_costs.xlsx')
```

```python
# Sense check the DataFrame
transport_costs.shape
```

```python
transport_costs.head()
```

### Subset the data

```python
# Subset data to exclude the Currency and Distance Unit columns
transport_totals = pd.read_excel('transport_costs.xlsx', usecols=['region', 'country', 'port', 'sea freight cost', 'road transport cost per km'])
```

```python
transport_totals.head()
```

### Add new columns to hold calculations

```python
# Add total sea freight cost column
transport_totals_all = pd.DataFrame(transport_totals, 
                                     columns=['region', 'country', 'port', 'sea freight cost', 
                                              'total sea freight cost', 'road transport cost per km', 
                                              'total road transport cost', 'total export costs'])
```

```python
# Calculate total sea freight cost
transport_totals_all['total sea freight cost'] = transport_totals_all['sea freight cost']*2*4
```

```python
# Calculate total road transport cost
transport_totals_all['total road transport cost'] = transport_totals_all['road transport cost per km']*10000
```

```python
# Calculate total export cost
transport_totals_all['total export costs'] = transport_totals_all['total sea freight cost']+transport_totals_all['total road transport cost']
```

Export data as a csv file

```python
transport_totals_all.to_csv('transport_totals.csv', index=False)
```
